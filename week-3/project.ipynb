```python
import tkinter as tk
from tkinter import messagebox
import pandas as pd

# 1. LOAD DATA (At the very top)
try:
    df = pd.read_csv('GIG-logistics(in).csv')
except FileNotFoundError:
    # Use a popup here so the user sees the error immediately
    messagebox.showerror("File Error", "GIG-logistics.csv not found in this folder!")

# 2. THE POPUP WINDOW (Define first)
def show_welcome_window(name, colleagues):
    welcome_win = tk.Toplevel(root)
    welcome_win.title("GIG Admin Box")
    welcome_win.geometry("350x400")

    tk.Label(welcome_win, text=f"Welcome, {name}!", font=("Arial", 12, "bold"), pady=10).pack()
    tk.Label(welcome_win, text="Your Department Members:", fg="blue").pack()

    if colleagues:
        for member in colleagues:
            tk.Label(welcome_win, text=f"• {member}").pack()
    else:
        tk.Label(welcome_win, text="No other members in this department.").pack()

# 3. THE SEARCH LOGIC (The Brain)
def check_employee():
    # Get values from the GUI boxes
    name_input = name_entry.get().strip()
    dept_input = dept_entry.get().strip()

    # Search (Case-Insensitive)
    # Ensure 'name' and 'Department' match your CSV headers exactly!
    match = df[(df['Name'].str.upper() == name_input.upper()) & 
               (df['Department'].str.upper() == dept_input.upper())]

    if not match.empty:
        # Find colleagues (Same dept, different name)
        colleagues = df[(df['Department'].str.upper() == dept_input.upper()) & 
                        (df['Name'].str.upper() != name_input.upper())]
        colleague_list = colleagues['name'].tolist()
        
        show_welcome_window(name_input, colleague_list)
    else:
        messagebox.showerror("Verification Failed", f"Employee '{name_input}' not found in {dept_input}.")

# 4. THE MAIN GUI
root = tk.Tk()
root.title("GIG Logistics - Employee Verification")
root.geometry("400x300")

tk.Label(root, text="Enter Name:").pack(pady=5)
name_entry = tk.Entry(root)
name_entry.pack()

tk.Label(root, text="Enter Department:").pack(pady=5)
dept_entry = tk.Entry(root)
dept_entry.pack()

submit_button = tk.Button(root, text="Verify Identity", command=check_employee)
submit_button.pack(pady=20)

# 5. START THE APP
root.mainloop()
```
