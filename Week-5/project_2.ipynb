```python
import tkinter as tk
from tkinter import messagebox

class PriceCalculator:
    def __init__(self):
        # This dictionary stores location and price
        self.pricing_data = {
            ("PAU", True): 2000,
            ("PAU", False): 1500,
            ("Epe", True): 5000,
            ("Epe", False): 4000
        }

    def get_price(self, location, weight):
        #Logic to determine price based on location and weight threshold.
        is_heavy = weight >= 10
        return self.pricing_data.get((location, is_heavy))

class DeliveryGUI:
    def __init__(self, window):
        self.calculator = PriceCalculator()
        self.window = window
        self.window.title("Delivery Price Calculator")
        self.window.geometry("350x250")

        # Location Selection
        tk.Label(window, text="Select Delivery Location:", font=("Arial", 10)).pack(pady=5)
        self.location_var = tk.StringVar(value="PAU")
        tk.OptionMenu(window, self.location_var, "PAU", "Epe").pack()

        # Weight Input
        tk.Label(window, text="Enter Package Weight (kg):", font=("Arial", 10)).pack(pady=5)
        self.weight_entry = tk.Entry(window)
        self.weight_entry.pack()

        # Calculate Button
        tk.Button(window, text="Calculate Fare", command=self.process_payment, bg="#007BFF", fg="white").pack(pady=20)

    def process_payment(self):
        try:
            loc = self.location_var.get()
            wgt = float(self.weight_entry.get())
            
            final_price = self.calculator.get_price(loc, wgt)
            
            if final_price:
                messagebox.showinfo("Delivery Quote", f"Location: {loc}\nWeight: {wgt}kg\n\nTotal to Pay: N{final_price}")
            else:
                messagebox.showerror("Error", "Selected location is not available.")
        except ValueError:
            messagebox.showerror("Input Error", "Please enter a valid number for the weight.")

if __name__ == "__main__":
    root = tk.Tk()
    app = DeliveryGUI(root)
    root.mainloop()
```
