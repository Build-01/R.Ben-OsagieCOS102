```python
import tkinter as tk
from tkinter import messagebox
import random

class Employee:
    def __init__(self):
        # Database of authorized staff
        self.staff_list = [
            "Mary Evans", "Eyo Ishan", "Durojaiye Dare", "Adams Ali", "Andrew Ugwu",
            "Stella Mankinde", "Jane Akibo", "Ago James", "Michell Taiwo", 
            "Abraham Jones", "Nicole Anide", "Kosi Korso", "Adele Martins", 
            "Emmanuel Ojo", "Ajayi Fatima"
        ]
        self.tasks = ["Loading", "Transporting", "Reviewing Orders", "Customer Service", "Delivering Items"]
        self.attendance = []

    def check_employee(self, name):
        #Verifies if the name exists in the staff list.
        if name in self.staff_list:
            self.take_attendance(name)
            assigned_task = self.assign_task()
            messagebox.showinfo("Access Granted", f"Welcome, {name}!\nAttendance marked.\nTask: {assigned_task}")
        else:
            self.refuse_access()

    def take_attendance(self, name):
        #Records the name in the attendance log.
        self.attendance.append(name)
        print(f"Log: {name} signed in.")

    def assign_task(self):
        #Randomly selects a task using the random module.
        return self.tasks[random.randint(0, len(self.tasks) - 1)]

    def refuse_access(self):
        #Politely informs the user they are not recognized.
        messagebox.showwarning("Access Denied", "We're sorry, your name is not in our system. Please contact Mrs. Jane.")

# --- GUI Setup ---
def start_app():
    system = Employee()
    root = tk.Tk()
    root.title("Jane's Logistics Portal")
    root.geometry("350x200")

    tk.Label(root, text="Employee Identity Verification", font=("Arial", 12, "bold")).pack(pady=10)
    tk.Label(root, text="Enter Full Name:").pack()
    
    entry = tk.Entry(root, width=30)
    entry.pack(pady=5)

    def process_login():
        user_name = entry.get().strip()
        system.check_employee(user_name)
        entry.delete(0, tk.END)

    tk.Button(root, text="Check Identity", command=process_login, bg="#4CAF50", fg="white").pack(pady=20)
    
    root.mainloop()

if __name__ == "__main__":
    start_app()
```
