```python
# Python program for calculating compound interest

# 1. Taking all inputs
# the abs keyword is to give the absolute value of whatever the user inputs to ensure no negative value(Error handling_1)
try:
    p = abs(float(input("Enter Principal: ")))
    r = abs(float(input("Enter rate of the interest: ")))
    t = abs(float(input("Enter time of the interest: ")))
    n = abs(float(input("Enter number of time: ")))

# Computing the values of amount and compound interest
    a = p * (1 + (r / 100) / n) ** (n * t)
    ci = a - p 
# formating the result to standard output(2.decimal places)
    print(f"The amount is: {a:,.2f}")
    print(f"The simple interest is: {ci:,.2f}")

# This runs if the user inputs letters (Error handling_2)
except ValueError:
    print("Error: Dear user, please enter numbers only.") 
# This ensures that the input for n is not zero
except ZeroDivisionError:
    print("Error: Dear user'Number of times' (n) cannot be zero.")
```
