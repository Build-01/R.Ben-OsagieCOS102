# Python program for calculating Annuity plan

try:
# 1. Taking all inputs
    pmt = abs(float(input("Please enter fixed payment amount: ")))
    r = abs(float(input("Please enter periodic interest rate: ")))
    t = abs(float(input("Please enter the time in years: ")))
    n = abs(float(input("Please enter number of payment period: ")))

# 2. Computing the values (Indented to stay inside 'try')
    if r == 0:
        a = pmt * n * t
    else:
        # Formula for Future Value of Ordinary Annuity
        a = pmt * ((1 + (r/100)/n) ** (n*t) - 1) / ((r/100)/n)
    
# 3. Formatting the result
    print(f"\nThe total future value is: {a:,.2f}")

except ValueError:
    print("Error: Dear user, please enter numbers only.") 
except ZeroDivisionError:
    print("Error: 'Number of times' (n) cannot be zero.")
