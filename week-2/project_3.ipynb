```python
# Python program to organize data into a table

import pandas as pd

# 1. Organize the data into lists (Combining Girls and Boys)
names = [
    "Evelyn", "Jessica", "Somto", "Edith", "Liza", "Madonna", "Waje", "Tola", "Aisha", "Latifa",
    "Chinedu", "Liam", "Wale", "Gbenga", "Abiola", "Kola", "Kunle", "George", "Thomas", "Wesley"
]

ages = [
    17, 16, 17, 18, 16, 18, 17, 20, 19, 17, # Girls
    19, 16, 18, 17, 20, 19, 16, 18, 17, 19  # Boys
]

heights = [
    5.5, 6.0, 5.4, 5.9, 5.6, 5.5, 6.1, 6.0, 5.7, 5.5, # Girls
    5.7, 5.9, 5.8, 6.1, 5.9, 5.5, 6.1, 5.4, 5.8, 5.7  # Boys
]

scores = [
    80, 85, 70, 60, 76, 66, 87, 95, 50, 49, # Girls
    74, 87, 75, 68, 66, 78, 87, 98, 54, 60  # Boys
]

# 2. Put the data into a dictionary
data_dict = {
    "Name": names,
    "Age": ages,
    "Height": heights,
    "Score": scores
}

# 3. Create the DataFrame (The Table)
df = pd.DataFrame(data_dict)

# 4. Display the table
print(df)
```
