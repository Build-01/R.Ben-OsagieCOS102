# Inheritance in Python

Inheritance is the capability of one class to derive or inherit the properties from another class. The benefits of inheritance are: 
 
<ol><li>
It represents real-world relationships well.</li><li>
It provides reusability of a code, so that you don’t have to write the same code again and again. </li><li>It allows for adding more features to a class without modifying it.</li><li>
It is transitive in nature, which means that if class B inherits from another class A, then all the subclasses of B would automatically inherit from class A.</li><ol>

### A Python program to demonstrate inheritance

```python
class Person(object):

    # Constructor
    def __init__(self, name):
        self.name = name

    # To get name method
    def getname(self):
        return self.name

    # To check if this person is an employee
    def isEmployee(self):
        return False
    

# Inherited or Subclass (Note Person in bracket)
class Employee(Person):

    # Here employee method returns true
    def isEmployee(self):
        return True

# Object Instantiation
emp = Person("Segun Da-Silver") # An Object of Person
print(emp.getname(), emp.isEmployee())

emp = Employee("Omotayo Ayeni") # An Object of Employee
print(emp.getname(), emp.isEmployee())
```

### What is object class? 
In Python (from version 3.x), object is root of all classes. 
In Python 3.x, <b>“class Test(object)”</b> and <b>“class Test”</b> are same. 

### Subclassing (Calling constructor of parent class) 
<ul><li>A child class needs to identify which class is its parent class.</li><li> This can be done by mentioning the parent class name in the definition of the child class. 
Eg: class subclass_name (superclass_name): </li></ul>

```python
# Python code to demonstrate how parent constructors
# are called.

# parent class
class Person(object):

        # __init__ is known as the constructor
        def __init__(self, name, idnumber):
            self.name = name
            self.idnumber = idnumber
            
        def display(self):
            print("Name:", self.name)
            print("ID No.:", self.idnumber)
            print("Salary:", self.salary)

# child class
class Employee(Person):
    def __init__(self, name, idnumber, salary, post, age):
        self.salary = salary
        self.post = post
    
        # invoking the __init__ of the parent class
        Person.__init__(self, name, idnumber)
   
                
# creation of an object variable or an instance
a = Employee('Adedoyin Adeniji', 886012, 200000, "Intern", 34)

# calling a function of the class Person using its instance
a.display()
```

### Note that:
<ul><li>In Python, every class inherits from a built-in basic class called ‘object’. The constructor i.e. the ‘__init__’ function of a class is invoked when we create an object variable or an instance of the class.</li><li>
The variables defined within __init__() are called as the instance variables or objects. Hence, ‘name’ and ‘idnumber’ are the objects of the class Person. Similarly, ‘salary’ and ‘post’ are the objects of the class Employee.</li><li> Since the class Employee inherits from class Person, ‘name’ and ‘idnumber’ are also the objects of class Employee.</li><li>
If you forget to invoke the __init__() of the parent class then its instance variables would not be available to the child class. </li></ul>

```python
# Python program to demonstrate error if we
# forget to invoke __init__() of the parent.

class A:
    # class attribute
    surname= "Moru"
    
    def __init__(self, n = 'Emmanuel Abbah'):
        #instance attribute
        self.name = n
        
class B(A):
    def __init__(self, roll):
        self.roll = roll
        # self.name = n

# Object Instance
project = B(23)
# print (project.name)
```

## Types of Inheritance in Python
There are four types of inheritance in Python. The type of Inheritance depends upon the number of child and parent classes involved. : 

#### Single Inheritance:
<ul><li> Single inheritance enables a derived class to inherit properties from a single parent class, thus enabling code reusability and the addition of new features to existing code.</li></ul>

```python
# Python program to demonstrate
# single inheritance


# Base class
class parent: 
    def func1(self):
        print("This function is in parent class.")

# Derived class
class child(parent):
    def func2(self):
        print("This function is in child class.")

# Object Instance
object = child()
object.func1()
object.func2()
```

#### Multiple Inheritance:
<ul><li>When a class can be derived from more than one base class this type of inheritance is called multiple inheritance. </li><li>In multiple inheritance, all the features of the base classes are inherited into the derived class. </li></ul>
 

```python
# Python program to demonstrate
# multiple inheritance


# Base class1
class Mother: 
    mothername = "Caroline Aina"
    def mother(self):
        print(self.mothername)

# Base class2
class Father:
    fathername = "Paul Abiodun"
    def father(self):
        print(sef.fathername)

# Derived class
class Son(Mother, Father):
    def parents(self):
        print("Father :", self.fathername)
        print("Mother :", self.mothername)

# Object Instance
s1 = Son()
s1.parents()
```

#### Multilevel Inheritance
<ul><li>In multilevel inheritance, features of the base class and the derived class are further inherited into the new derived class.</li><li> This is similar to a relationship representing a child and grandfather.</li></ul> 

```python
# Python program to demonstrate
# multilevel inheritance

# Base class
class Grandfather:

    def __init__(self, grandfathername):
        self.grandfathername = grandfathername

# Intermediate class
class Father(Grandfather):
    def __init__(self, fathername, grandfathername):
        self.fathername = fathername
        

        # invoking constructor of Grandfather class
        Grandfather.__init__(self, grandfathername)

# Derived class
class Son(Father):
    def __init__(self,sonname, fathername, grandfathername):
        self.sonname = sonname

        # invoking constructor of Father class
        Father.__init__(self, fathername, grandfathername)

    def print_name(self):
        print('Grandfather name :', self.grandfathername)
        print("Father name :", self.fathername)
        print("Son name :", self.sonname)

# Object Instance
s1 = Son('Gbenga', 'Oluwafemi', 'Adeoke')
print(s1.grandfathername)
s1.print_name()
```

#### Hierarchical Inheritance:
<ul><li>When more than one derived classes are created from a single base this type of inheritance is called hierarchical inheritance.</li><li> In this program, we have a parent (base) class and two child (derived) classes.</li></ul>

```python
# Python program to demonstrate
# Hierarchical inheritance


# Base class
class Parents:
    def func1(self):
        print("This function is in parent class.")

# Derived class1
class Child1(Parents):
    def func2(self):
        print("This function is in child 1.")

# Derivied class2
class Child2(Parents):
    def func3(self):
        print("This function is in child 2.")

# Object Instantiation
object1 = Child1()
object2 = Child2()
object1.func1()
object1.func2()
object2.func1()
object2.func3()
```

#### Hybrid Inheritance:
<ul><li>Inheritance consisting of multiple types of inheritance is called hybrid inheritance.</li></ul>

```python
# Python program to demonstrate
# hybrid inheritance


class School():
    def func1(self):
        print("This function is in school.")

class Student1(School):
    def func2(self):
        print("This function is in student 1. ")

class Student2(School):
    def func3(self):
        print("This function is in student 2.")

class Student3(Student1, School):
    def func4(self):
        print("This function is in student 3.")

# Object Instantiation
object = Student3()
object.func1()
object.func2()
```

# Polymorphism in Python

<ul><li>In programming, Polymorphism is a concept of OOP.</li><li>The word polymorphism means having many forms.</li><li>
    It enables using a single interface with the input of different data types, different classes or maybe for a different number of inputs.</li><li>Polymorphism means the same function name (but different signatures) being used for different types.</li></ul>

### Inbuilt polymorphic functions

```python
# Python program to demonstrate in-built polymorphic functions

# len() being used for a string
print(len("Department of Computer Science"))

# len() being used for a list
print(len([10, 20, 30, 40, 50, 60, 70]))
```

### User-defined polymorphic functions : 

```python
# A simple Python function to demonstrate
# Polymorphism

def add(x, y, z= 0, w=6): 
    return x + y  

# Driver code
print(add(2, 3))
print(add(2, 3, 4))
print(add(2, 3, 5, 6))
```

### Polymorphism with class methods: 
<ul><li>The code below shows how Python can use two different class types, in the same way.</li><li> We create a for loop that iterates through a tuple of objects. Then call the methods without being concerned about which class type each object is. We assume that these methods actually exist in each class.</li></ul> 

```python
class Nigeria():
    def Capital(self):
        print("Lagos is the capital of Nigeria.")

    def languages(self):
        print("English is the official language of Nigeria.")

    def type(self):
        print("Nigeria is the giant of Africa.")

class Togo():
    def Capital(self):
        print("Lome is the capital of Togo.")

    def languages(self):
        print("French is the primary language of Togo.")

    def type(self):
        print("Togo is known for its palm-lined beaches and hilltop villages.")

obj_naija = Nigeria()
obj_togo = Togo()
for country in (obj_naija, obj_togo):
    country.Capital()
    country.languages()
    country.type()
```

### Polymorphism with Inheritance:
<ul><li>In Python, Polymorphism lets us define methods in the child class that have the same name as the methods in the parent class. In inheritance, the child class inherits the methods from the parent class.</li><li> However, it is possible to modify a method in a child class that it has inherited from the parent class.</li><li> This is particularly useful in cases where the method inherited from the parent class doesn’t quite fit the child class. In such cases, we re-implement the method in the child class.</li><li> This process of re-implementing a method in the child class is known as <b>Method Overriding. </b> </li></ul>

```python
class Bird:
    def intro(self):
        print("There are many types of birds.")

    def flight(self):
        print("Most of the birds can fly but some cannot.")

class sparrow(Bird):
    def flight(self):
        print("Sparrows can fly.")

class ostrich(Bird):
    def flight(selfie):
        print("Ostriches cannot fly.")

obj_bird = Bird()
obj_spr = sparrow()
obj_ost = ostrich()

obj_bird.intro()
obj_bird.flight()

obj_spr.intro()
obj_spr.flight()

obj_ost.intro()
obj_ost.flight()
```

Mary, Agatha and Noel work with Zenith Bank Nigeria. Mary works in the Retail Banking Division, Agatha works in the Global Banking Division and Noel works in the Commercial Banking Division. The three divisions have some unique services and some mutual services as indicated:

#### Retail Banking:

With your knowledge in OOP develop a python GUI program that will take as input an employee name and division, and then displays the service rendered in the division. The program should highlght key concepts of OOP; class objects inheritance and polymorphism.

#### Hints:
<ul><li>Create parent class <b>zenith()</b> with two methods <b>unique_services()</b> and <b>mutual_services()</b></li><li>
    The different divisions can be subclasses of the parent class, inheriting the parent methods.</li><li>
    Ploymorphism can be used to overide exclusive services.</li><ul>

```python
import tkinter as tk
from tkinter import ttk, messagebox
# Still going to process the knowledge gain from this work.
# I had to show up for weekly tasks but have dry coding tomorrow, i'm guilty of working with AI.
# OOP CLASSES

class Zenith:
    #Parent class representing Zenith Bank division
    
    def __init__(self, division_name):
        self.division_name = division_name
    
    def unique_services(self):
        #Polymorphic method - to be overridden by subclasses
        return []
    
    def mutual_services(self):
        #Common mutual services across all divisions (can be overridden)
        return []
    
    def get_all_services(self):
        #Combines mutual and unique services
        mutual = self.mutual_services()
        unique = self.unique_services()
        return mutual + unique


class RetailBanking(Zenith):
    #Retail Banking Division - Subclass of Zenith
    
    def unique_services(self):
        #Override: Retail-specific services
        return [
            "Loans and Mortgages",
            "Checking and Savings Accounts",
            "Retirement and Education Accounts"
        ]
    
    def mutual_services(self):
        #Override: Mutual services offered by Retail (shared with Commercial)
        return [
            "Lines of Credit",
            "Investment Management and Accounts",
            "Insurance"
        ]


class GlobalBanking(Zenith):
    #Global Banking Division - Subclass of Zenith
    
    def unique_services(self):
        #Override: Global-specific services
        return [
            "Multi-currency Management Services and Products",
            "Foreign Currency Accounts",
            "Foreign Currency Credit Cards",
            "Transborder Advisory Services",
            "Liquidity Management"
        ]
    
    def mutual_services(self):
        #Global Banking has NO mutual services with others in this context
        return []


class CommercialBanking(Zenith):
    #Commercial Banking Division - Subclass of Zenith
    
    def unique_services(self):
        #Override: Commercial-specific services
        return [
            "Advisory Services"
        ]
    
    def mutual_services(self):
        #Override: Mutual services offered by Commercial (shared with Retail)
        return [
            "Lines of Credit",
            "Investment Management and Accounts",
            "Insurance"
        ]


# ---------------------- GUI APPLICATION ----------------------

class ZenithBankApp:
    #Main GUI application class
    
    def __init__(self, root):
        self.root = root
        self.root.title("Zenith Bank Nigeria - Division Services")
        self.root.geometry("750x600")
        self.root.resizable(False, False)
        
        # Color scheme
        self.bg_color = "#0A2540"      # Dark blue (Zenith brand)
        self.primary_color = "#00A3E0"  # Light blue
        self.secondary_color = "#F5F7FA"
        self.accent_color = "#27AE60"   # Green accent
        
        # Configure root window background
        self.root.configure(bg=self.bg_color)
        
        # Set custom styles for ttk widgets
        self.setup_styles()
        
        # Create GUI elements
        self.create_header()
        self.create_input_frame()
        self.create_output_frame()
        
        # Initialize division mapping
        self.division_map = {
            "Retail Banking": RetailBanking,
            "Global Banking": GlobalBanking,
            "Commercial Banking": CommercialBanking
        }
    
    def setup_styles(self):
        #Configure ttk styles for modern look
        style = ttk.Style()
        style.theme_use('clam')
        
        # Configure Heading style
        style.configure("Heading.TLabel", 
                        font=("Segoe UI", 18, "bold"),
                        foreground="white",
                        background=self.bg_color)
        
        # Configure Subheading style
        style.configure("Subheading.TLabel",
                        font=("Segoe UI", 10),
                        foreground="#CCCCCC",
                        background=self.bg_color)
        
        # Configure Input label style
        style.configure("Input.TLabel",
                        font=("Segoe UI", 11, "bold"),
                        foreground="white",
                        background=self.bg_color)
        
        # Configure Button style
        style.configure("Primary.TButton",
                        font=("Segoe UI", 11, "bold"),
                        background=self.primary_color,
                        foreground="white",
                        padding=10)
        style.map("Primary.TButton",
                  background=[('active', '#0082B0')])
        
        # Configure Frame style
        style.configure("Card.TFrame",
                        background="white",
                        relief="flat",
                        borderwidth=0)
        
        # Configure Output label style
        style.configure("OutputTitle.TLabel",
                        font=("Segoe UI", 13, "bold"),
                        foreground="#0A2540",
                        background="white")
        
        style.configure("Service.TLabel",
                        font=("Segoe UI", 10),
                        foreground="#333333",
                        background="white",
                        wraplength=600)
        
        # Configure Combobox style
        self.root.option_add('*TCombobox*Listbox.font', ("Segoe UI", 10))
    
    def create_header(self):
        #Create header section with title
        header_frame = tk.Frame(self.root, bg=self.bg_color, height=120)
        header_frame.pack(fill="x", padx=20, pady=(20, 10))
        
        # Title
        title_label = ttk.Label(header_frame, text="ZENITH BANK NIGERIA", 
                                style="Heading.TLabel")
        title_label.pack(pady=(10, 5))
        
        # Subtitle
        subtitle_label = ttk.Label(header_frame, 
                                   text="Division Services Information System",
                                   style="Subheading.TLabel")
        subtitle_label.pack()
        
        # Divider line
        divider = tk.Frame(self.root, bg=self.primary_color, height=2)
        divider.pack(fill="x", padx=20)
    
    def create_input_frame(self):
        #Create input section with employee details
        input_container = tk.Frame(self.root, bg=self.bg_color)
        input_container.pack(fill="x", padx=20, pady=20)
        
        # Card-like frame for input
        card_frame = tk.Frame(input_container, bg="white", 
                              relief="flat", bd=0, highlightthickness=1,
                              highlightcolor="#DDDDDD", highlightbackground="#DDDDDD")
        card_frame.pack(fill="x", padx=10, pady=10)
        
        # Inner padding frame
        inner_frame = tk.Frame(card_frame, bg="white")
        inner_frame.pack(padx=25, pady=25)
        
        # Employee Name
        name_label = ttk.Label(inner_frame, text="Employee Name:", 
                               style="Input.TLabel", background="white",
                               foreground="#0A2540")
        name_label.grid(row=0, column=0, sticky="w", pady=(0, 5))
        
        self.name_entry = tk.Entry(inner_frame, font=("Segoe UI", 11),
                                   relief="solid", bd=1, highlightthickness=0)
        self.name_entry.grid(row=1, column=0, sticky="ew", pady=(0, 15))
        self.name_entry.configure(bg="#F8F9FA", fg="#333333")
        
        # Division
        div_label = ttk.Label(inner_frame, text="Division:", 
                              style="Input.TLabel", background="white",
                              foreground="#0A2540")
        div_label.grid(row=2, column=0, sticky="w", pady=(0, 5))
        
        self.division_var = tk.StringVar()
        self.division_combo = ttk.Combobox(inner_frame, textvariable=self.division_var,
                                           values=["Retail Banking", "Global Banking", "Commercial Banking"],
                                           font=("Segoe UI", 11), state="readonly")
        self.division_combo.grid(row=3, column=0, sticky="ew", pady=(0, 20))
        self.division_combo.set("Select Division")
        
        # Submit Button
        self.submit_btn = ttk.Button(inner_frame, text="SHOW SERVICES", 
                                     command=self.display_services,
                                     style="Primary.TButton")
        self.submit_btn.grid(row=4, column=0, pady=(0, 10))
        
        # Clear Button
        self.clear_btn = tk.Button(inner_frame, text="Clear", 
                                   command=self.clear_all,
                                   font=("Segoe UI", 10),
                                   bg="#E74C3C", fg="white",
                                   relief="flat", padx=20, pady=5,
                                   activebackground="#C0392B",
                                   activeforeground="white",
                                   cursor="hand2")
        self.clear_btn.grid(row=5, column=0)
        
        # Configure grid column weight
        inner_frame.columnconfigure(0, weight=1)
    
    def create_output_frame(self):
        #Create output section for displaying services
        output_container = tk.Frame(self.root, bg=self.bg_color)
        output_container.pack(fill="both", expand=True, padx=20, pady=(0, 20))
        
        # Card-like frame for output
        self.output_card = tk.Frame(output_container, bg="white",
                                    relief="flat", bd=0, highlightthickness=1,
                                    highlightcolor="#DDDDDD", highlightbackground="#DDDDDD")
        self.output_card.pack(fill="both", expand=True, padx=10, pady=10)
        
        # Inner padding frame
        self.output_inner = tk.Frame(self.output_card, bg="white")
        self.output_inner.pack(fill="both", expand=True, padx=25, pady=20)
        
        # Initial placeholder message
        self.show_placeholder()
    
    def show_placeholder(self):
        #Show placeholder message when no division is selected
        # Clear existing widgets
        for widget in self.output_inner.winfo_children():
            widget.destroy()
        
        placeholder_label = tk.Label(self.output_inner,
                                     text="Select a division and click 'SHOW SERVICES' to view available services",
                                     font=("Segoe UI", 11, "italic"),
                                     fg="#999999", bg="white",
                                     wraplength=550)
        placeholder_label.pack(expand=True, pady=40)
    
    def display_services(self):
        #Main method to fetch and display services based on selected division
        # Get input values
        employee_name = self.name_entry.get().strip()
        division = self.division_var.get()
        
        # Validation
        if not employee_name:
            messagebox.showwarning("Input Error", "Please enter the employee name.")
            return
        
        if division == "Select Division" or not division:
            messagebox.showwarning("Input Error", "Please select a division.")
            return
        
        # Create appropriate division object (Polymorphism in action)
        division_class = self.division_map.get(division)
        if not division_class:
            messagebox.showerror("Error", "Invalid division selected.")
            return
        
        # Instantiate the division object
        division_obj = division_class(division)
        
        # Get services using polymorphism
        unique_services = division_obj.unique_services()
        mutual_services = division_obj.mutual_services()
        
        # Display the services
        self.show_services(employee_name, division, mutual_services, unique_services)
    
    def show_services(self, name, division, mutual, unique):
        #Display services in the output frame with beautiful formatting
        # Clear existing widgets
        for widget in self.output_inner.winfo_children():
            widget.destroy()
        
        # Header with employee info
        header_text = f"Employee: {name.upper()}  |  Division: {division.upper()}"
        header_label = tk.Label(self.output_inner, text=header_text,
                                font=("Segoe UI", 12, "bold"),
                                fg=self.primary_color, bg="white")
        header_label.pack(anchor="w", pady=(0, 15))
        
        # Separator
        separator = tk.Frame(self.output_inner, bg=self.primary_color, height=2)
        separator.pack(fill="x", pady=(0, 15))
        
        # Debug print to console
        print(f"Mutual services: {mutual}")
        print(f"Unique services: {unique}")
        
        # Mutual Services Section
        if mutual and len(mutual) > 0:
            mutual_title = tk.Label(self.output_inner, text="MUTUAL SERVICES (Shared with other divisions)",
                                    font=("Segoe UI", 11, "bold"),
                                    fg="#27AE60", bg="white")
            mutual_title.pack(anchor="w", pady=(0, 10))
            
            for service in mutual:
                service_label = tk.Label(self.output_inner, text=f"  • {service}",
                                         font=("Segoe UI", 10),
                                         fg="#555555", bg="white",
                                         anchor="w", justify="left")
                service_label.pack(anchor="w", pady=2)
            
            if unique and len(unique) > 0:
                # Spacer
                tk.Frame(self.output_inner, bg="white", height=10).pack()
        else:
            # Show message if no mutual services
            no_mutual = tk.Label(self.output_inner, text="No mutual services shared with other divisions",
                                font=("Segoe UI", 10, "italic"),
                                fg="#999999", bg="white")
            no_mutual.pack(anchor="w", pady=(0, 10))
        
        # Unique Services Section
        if unique and len(unique) > 0:
            unique_title = tk.Label(self.output_inner, text="UNIQUE SERVICES (Exclusive to this division)",
                                    font=("Segoe UI", 11, "bold"),
                                    fg="#E67E22", bg="white")
            unique_title.pack(anchor="w", pady=(0, 10))
            
            for service in unique:
                service_label = tk.Label(self.output_inner, text=f"  ★ {service}",
                                         font=("Segoe UI", 10),
                                         fg="#333333", bg="white",
                                         anchor="w", justify="left")
                service_label.pack(anchor="w", pady=2)
        else:
            # Show message if no unique services
            no_unique = tk.Label(self.output_inner, text="No exclusive services for this division",
                                font=("Segoe UI", 10, "italic"),
                                fg="#999999", bg="white")
            no_unique.pack(anchor="w", pady=(0, 10))
        
        # Footer
        tk.Frame(self.output_inner, bg="white", height=10).pack()
        footer = tk.Label(self.output_inner, text="— Zenith Bank Nigeria —",
                          font=("Segoe UI", 8, "italic"),
                          fg="#AAAAAA", bg="white")
        footer.pack(side="bottom", pady=(15, 0))
    
    def clear_all(self):
        #Clear all input fields and reset output display
        self.name_entry.delete(0, tk.END)
        self.division_var.set("Select Division")
        self.show_placeholder()


# ---------------------- MAIN ENTRY POINT ----------------------

if __name__ == "__main__":
    root = tk.Tk()
    app = ZenithBankApp(root)
    root.mainloop()
```
